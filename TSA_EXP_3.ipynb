```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf

# Time Series Plot
def plot_time_series(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data)
    plt.title('Time Series Plot')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.show()

# Autocorrelation Plot
def plot_acf_series(data):
    plt.figure(figsize=(10, 6))
    plot_acf(data, lags=40)
    plt.title('Autocorrelation Plot')
    plt.show()

# Augmented Dickey-Fuller (ADF) Test for Stationarity
def adf_test(data):
    result = adfuller(data)
    print("ADF Statistic: %f" % result[0])
    print("p-value: %f" % result[1])

    if result[1] <= 0.05:
        print("The series is stationary (reject null hypothesis)")
    else:
        print("The series is non-stationary (fail to reject null hypothesis)")

# KPSS Test for Stationarity (Trend Stationarity)
def kpss_test(data):
    result = kpss(data, regression='c')
    print("KPSS Statistic: %f" % result[0])
    print("p-value: %f" % result[1])

    if result[1] <= 0.05:
        print("The series is non-stationary (reject null hypothesis)")
    else:
        print("The series is stationary (fail to reject null hypothesis)")
```

```python
# Example 1: Generate synthetic random walk data (non-stationary)
np.random.seed(42)
time_series_data = pd.Series(np.random.randn(100).cumsum())

# Example 2: Simulate a stationary random series
stationary_data = pd.Series(np.random.randn(100))

# Example 3: Create a trend-based non-stationary series
non_stationary_trend = pd.Series(np.arange(100) + np.random.randn(100))
```

```python
# Test on the non-stationary data (random walk)
print("Testing for non-stationary random walk series:")
plot_time_series(time_series_data)
plot_acf_series(time_series_data)

print("\nPerforming ADF Test on random walk:")
adf_test(time_series_data)

print("\nPerforming KPSS Test on random walk:")
kpss_test(time_series_data)

# Test on stationary data
print("\nTesting for stationary random series:")
plot_time_series(stationary_data)
plot_acf_series(stationary_data)

print("\nPerforming ADF Test on stationary series:")
adf_test(stationary_data)

print("\nPerforming KPSS Test on stationary series:")
kpss_test(stationary_data)

# Test on trend-based non-stationary data
print("\nTesting for trend-based non-stationary series:")
plot_time_series(non_stationary_trend)
plot_acf_series(non_stationary_trend)

print("\nPerforming ADF Test on trend-based non-stationary series:")
adf_test(non_stationary_trend)

print("\nPerforming KPSS Test on trend-based non-stationary series:")
kpss_test(non_stationary_trend)
```
